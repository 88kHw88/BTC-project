```python
from pprint import pprint
```

```python
text = """
For some quick analysis, creating a corpus could be overkill.
If all you need is a word list,
there are simpler ways to achieve that goal."""
```

```python
pprint(nltk.word_tokenize(text), width=75, compact=True)
```

```python
from __future__ import annotations
from typing import List

#Type hints
#https://realpython.com/lessons/type-hinting/ 

#variable: type hints = nltk.word_tokenize(text)
#Type hinting gives you the ability to define the type of 
#values that can be passed for each argument of a function or method 

#the tokenized words of (text) are set in list with 'str' stored in variable 'woord'
woord: list[str] = nltk.word_tokenize(text)
#.FreqDist counts the words in (text)
fd = nltk.FreqDist(woord)
#apply tabulate for overview
fd.tabulate(10)
#apply most_comman to count the common
fd.most_common(5)
```

```python
lower_fd = nltk.FreqDist([w.lower() for w in fd])
lower_fd
```

```python
#nltk.vocab
zin: list[str] = nltk.word_tokenize("""
Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.""")

text = nltk.Text(zin)
fd = text.vocab()
fd
```

```python
words = [w for w in nltk.corpus.state_union.words() if w.isalpha()]
finder = nltk.collocations.TrigramCollocationFinder.from_words(words)

```

```python
finder.ngram_fd.most_common(2)
finder.ngram_fd.tabulate(2)
```
